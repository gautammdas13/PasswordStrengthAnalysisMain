# Abstract

In an era of increasing cybersecurity threats, understanding password vulnerabilities is crucial to improving online security. This study investigates how different password categories—such as alphanumeric combinations, animal names, and personal names—impact password strength and resistance to brute-force attacks. Our goal was to analyze how these categories influence the time required for an attacker to crack a password and whether combining multiple categories enhances security.

To conduct this research, we gathered a dataset of passwords categorized by type and applied brute-force attack simulations using common password-cracking techniques. By measuring the time complexity and strength of these attacks, we assessed the relative security of each password category. Statistical analyses (specifically T-tests, ANOVA, and Tukey HSD) were performed to determine the effectiveness of different password structures in resisting attacks.

Our findings indicate that purely alphanumeric passwords, particularly those with randomized characters, offer the highest resistance to brute-force attacks. Conversely, passwords composed of common words, names, or predictable patterns are significantly more vulnerable. However, we observed that hybrid passwords—those incorporating multiple categories—demonstrate improved security compared to single-category passwords.

These results highlight the importance of password composition in cybersecurity. Our research suggests that users should avoid predictable patterns and instead utilize a mix of character types to maximize security. These insights can inform password policy recommendations, encouraging the adoption of stronger authentication practices to mitigate the risk of unauthorized access.

### __Time to crack (https://www.grc.com/haystack.htm)__

This webpage contains a numerical way of calculating the time to crack for an inputted password. When typing in the password into the input bar, the webpage would calculate the following categories for the value:
|Brute Force Search Space Analysis|Time Required to Exhaustively Search this Password's Space|
|---|---|
|Search Space Depth (Alphabet)|Online Attack Scenario(Assuming one thousand guesses per second)|
|Search Space Length (Characters)|Offline Fast Attack Scenario(Assuming one hundred billion guesses per second)|
|Exact Search Space Size (Count)(count of all possible passwords with this alphabet size and up to this password's length)|Massive Cracking Array Scenario (Assuming one hundred trillion guesses per second)|
|Search Space Size (as a power of 10)||

Each of these categories would be tabulated for the password and give us according values. From these values we will specifically use the Online Attack Scenario and the Offline Fast Attack Scenario in order to accurately represent real world threats to these passwords. The Online attack represents your everday brute force attempt to crack passwords. With many websites having limitations on password entry rates and other authenticating features like reCaptchas, the one thousand guesses per second represented by this scenario served as a good sweet spot. Now in order to represent some of the more advanced password cracking events like ones in which hackers have gained access to a database breach, the Offline Fast Attack is a perfect representation. The reason we chose to disregard the Massive Cracking Array is because the levels of computational power (quantum computing) needed to carry out one of these attacks is barely possible for some of the most sophisticated and powerful quantum computing clusters can handle. As such it would only skew our data to an highly-impossible situation and taking our analysis away from real-world/practical applications.

#### Define the strength webscraping algorithm and calculations

```python
# Initialize the WebDriver 

driver = webdriver.Chrome()
driver.get("https://www.grc.com/haystack.htm")
time.sleep(2)  

#Function to parse string and return value as float for Data Analysis

def parse_time_string(time_str: str) -> float:
    if not time_str or "error" in time_str.lower():
        return None
    
    # Regex pattern that captures a numeric value and a time unit

    pattern = r"([\d\.]+)\s*(seconds|minutes|hours|days|weeks|months|years|centuries)"
    match = re.search(pattern, time_str.lower())
    if not match:
        return None

    value = float(match.group(1))
    unit = match.group(2)
    
    if unit == "seconds":
        return value
    elif unit == "minutes":
        return value * 60
    elif unit == "hours":
        return value * 3600
    elif unit == "days":
        return value * 86400
    elif unit == "weeks":
        return value * 604800
    elif unit == "months":
        return value * 2629800
    elif unit == "years":
        return value * 31536000
    elif unit == "centuries":
        return value * 31536000 * 100 
    return None

def get_time_to_crack(password: str):
    try:
        # Wait for the password input field to be present and input the password

        password_input = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, 'passcode'))
        )
        password_input.clear()
        password_input.send_keys(password)
        
        # Allow time for the page to update the results

        time.sleep(1)
        
        # Locate the elements for Online and Offline attack times

        online_elem = driver.find_element(By.ID, 'online')
        offline_elem = driver.find_element(By.ID, 'offline')
        
        raw_online = online_elem.text.strip()
        raw_offline = offline_elem.text.strip()
        
        # Parse the raw strings into numeric seconds

        online_time = parse_time_string(raw_online)
        offline_time = parse_time_string(raw_offline)
        
        return online_time, offline_time
    
    except Exception as e:
        print(f"Error processing '{password}': {e}")
        return None, None

# Create a new DataFrame to store results

time_results = []
for password in passwords['password']:
    print(f"Processing password: {password}")
    online_time, offline_time = get_time_to_crack(password)
    time_results.append({
        'password': password,
        'online_time': online_time,
        'offline_time': offline_time
    })

time_df = pd.DataFrame(time_results)
print("Updated time_df with time-to-crack values (in seconds):")
display(time_df.head())

# Close the WebDriver

driver.quit()
```

### __Strength (https://www.uic.edu/apps/strong-password/)__

This webpage contains a numerical system of calculating the stregnth of an inputted password. Each of these values are tabulated when the password is inputted. The webpage would scan the password for the following categories within the password. Each of these values is followed by a formula for how they are calculated, with some having a higher "rate" than others:
|Category|Rate|
|---|---|
|Number of characters|+(n*4)|
|Uppercase letters|+((len-n)*2)|
|Lowercase Letters|+((len-n)*2)|
|Numbers|+(n*4)|
|Symbols|+(n*6)|
|Middle numbers|+(n*2)|
|Requirements|+(n*2)|

With these values, the total "rate" of the password is calculated as a subtotal. Then the webpage would calcuate the password for possible "deducations." These cases are important to finding the strength of the password since they are typically found in dictionary attack and brute force methods, which we are investigating. Here are the following categories and their according "rates" at which the password deducations are calculated:
|Category|Deducation Rate|
|---|---|
|Letters only|-n |	
|Numbers only|-n |	
| Repeat Characters (case insensitive)|(too complicated of an equation, truncated)|
|Consecutive uppercase letters|-(n*2) 	|
|Consecutive lowercase letters|-(n*2) 	|
|Consecutive numbers| -(n*2) 	|
|Sequential letters (3+)|-(n*3) 	|
| Sequential numbers (3+)|-(n*3) 	|
|Sequential symbols (3+)| -(n*3)|

With these two subtotals, the webpage would generate a subtotal for which we can tabulate and label each password with. This method would allow each password to be categorized, by strength, under the same method. This would ensure that there are no outstanding outliers or passwords that have "special strengths" compared to others. In addition, this would allow us to focus on the password categories (our research question) rather than dealing with unexplainable outliers in their strength.

> More information about the method of calculation on the webpage.

#### Define the strength webscraping algorithm and calculations

```python
# Initialize the WebDriver (utilizing Chrome)

driver = webdriver.Chrome()

# Open the page

driver.get('https://www.uic.edu/apps/strong-password/')

# Function to get bonus values for a password

def get_bonus_values(password):
    try:
        # Wait for the password input field to be present

        password_input = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, 'passwordPwd'))
        )
        
        # Clear and enter the password

        password_input.clear()
        password_input.send_keys(password)
        
        # Wait for results to update (2 seconds should suffice, adjust if needed)

        time.sleep(2)

        # There is a table where the values are calculated, we must find and store these values. We are most interested in the bonus values,
        # these values contain the numerical values calculated from the categories that the passwords fall under. These values are specifically
        # boxPlus and boxMinus. These values are extracted, as strings, and converted into integers, then tabulated to calcualte the
        # strength measurement of the password.

        # Extract boxPlus (bonus values)

        box_plus_elements = driver.find_elements(By.CLASS_NAME, 'boxPlus')
        box_plus_values = [int(element.text.strip().replace('+', '')) for element in box_plus_elements]

        # Extract boxMinus (deduction values)

        box_minus_elements = driver.find_elements(By.CLASS_NAME, 'boxMinus')
        box_minus_values = [int(element.text.strip().replace('-', '')) for element in box_minus_elements]

        return box_plus_values, box_minus_values

    except Exception as e:
        print(f"Error processing password '{password}': {e}")
        return [], []

# List to store results

strength_results = []

# Loop over each password in the dataframe and get the bonus/deduction values

for password in passwords['password']:
    print(f"Processing password: {password}")
    box_plus_values, box_minus_values = get_bonus_values(password)
    
    # Calculate the strength by summing boxPlus values and subtracting boxMinus values

    total_box_plus = sum(box_plus_values)
    total_box_minus = sum(box_minus_values)
    strength = total_box_plus - total_box_minus
    
    strength_results.append({
        'password': password,
        'strength': strength
    })

# Convert the results into a new DataFrame

results_df = pd.DataFrame(strength_results)

# Show the results

print(results_df)

# Close the WebDriver

driver.quit()
```

#### Detecting Outliers

We will be using Z-score to measure outliers. Within this analysis, using Z-score is useful due to how most of the data is similar to a Gaussian curve, with a common mean or average (which we will be able to see in later histograms). By utilizing Z-score, we can visually see and remove values which are outside of the bounds that we set for the dataset. For our case, we set the bounds to be at least 3 standard deviations away from the mean. This would allow us to get rid of passwords that could possibly influence our analysis due to its skewness towards certain values within the categories. 

```python
# Function to detect outliers using Z-Score

def detect_outliers_zscore(data, column, threshold=3):
    z_scores = np.abs(zscore(data[column]))
    return data[z_scores > threshold]
```

#### Removing Outliers

```python
# Function to remove outliers using Z-score method

def remove_outliers_zscore(data, columns, threshold=3):
    cleaned_data = data.copy()
    for col in columns:
        z_scores = np.abs(zscore(cleaned_data[col]))
        cleaned_data = cleaned_data[z_scores < threshold]
    return cleaned_data
```

```python
# List of numeric columns to clean

numeric_columns = ['online_time', 'offline_time', 'strength']
```

We want to find whether there is actually a statistical difference between the categories and their crack times and also their strengths. This would allow us to answer our question on whether the having specific password categories have higher crack times and strengths. 

### Correlations

```python
# Load the dataset

df = pd.read_csv("./Data/passwords_full_cleaned.csv")
```

We have to calculate the percent difference of the categories and their influence on the tested variables (the different categories) and the crack time of the passwords. This will also allow us to determine whether our hypothesis and statistical inferences/analysis is valid.

```python
# Function to calculate percentage difference

def percent_difference(group1, group2):
    return (group1.mean() - group2.mean()) / group2.mean() * 100
```

#### T-tests

We are proceeding with T-tests to see if there is a difference between them and the categories of the passwords. Since we are assuming that the data is normally distributed (as shown earlier), and there isn't too much variation, this would allow us to carry out this test. This would allow us to determine whether using time to crack values is valid in terms of password categories. Moreover, it would allow us to determine is our analysis displays results that are statistically valid.

```python
# T-tests: Name-based vs Numeric/Special Character Passwords

name_based = df[df['category'] == 'name']
numeric_special = df[df['category'].isin(['simple-alphanumeric', 'password-related'])]

t_online, p_online = stats.ttest_ind(name_based['online_time'], numeric_special['online_time'], equal_var=False)
t_offline, p_offline = stats.ttest_ind(name_based['offline_time'], numeric_special['offline_time'], equal_var=False)
```

```python
# Calculate the actual percentage difference in cracking time

online_diff = percent_difference(name_based['online_time'], numeric_special['online_time'])
offline_diff = percent_difference(name_based['offline_time'], numeric_special['offline_time'])

print(f"\nT-Test: Name vs Numeric/Special Characters (Online Time) → t = {t_online:.4f}, p = {p_online:.4f}, Diff = {online_diff:.2f}%")
print(f"T-Test: Name vs Numeric/Special Characters (Offline Time) → t = {t_offline:.4f}, p = {p_offline:.4f}, Diff = {offline_diff:.2f}%")
```

```python
# Compute Cohen's d effect size
mean_diff = np.mean(name_based['online_time']) - np.mean(numeric_special['online_time'])
pooled_std = np.sqrt((np.std(name_based['online_time'], ddof=1) ** 2 + np.std(numeric_special['online_time'], ddof=1) ** 2) / 2)
cohen_d = mean_diff / pooled_std
print(f'Cohen’s d for Name vs Numeric/Special Characters (Online Time): {cohen_d:.3f}')

mean_diff = np.mean(name_based['offline_time']) - np.mean(numeric_special['offline_time'])
pooled_std = np.sqrt((np.std(name_based['offline_time'], ddof=1) ** 2 + np.std(numeric_special['offline_time'], ddof=1) ** 2) / 2)
cohen_d = mean_diff / pooled_std
print(f'Cohen’s d for Name vs Numeric/Special Characters (Offline Time): {cohen_d:.3f}')
```

For both the offline and online times the t-test is a small value which shows that there is little difference between Name and Numeric/Special Characters when it comes to time to crack. The p-value for both is greater than 0.05 indicating the difference between Name and Numeric/Special Characters is not statistically significant. As for the t-value, this value would allow us to see the difference between the groups means in relation to variability, similar to standard deviation. In other words, it allows us to see the deviation of the mixed and non-mixed categories in relation to the crack times. Looking at this result, there is some variation, around ~1.023 t deviations. Finally, the percent difference value shows us that there is a difference between the two categories in terms of the cracking time (around ~49.5%), restating what the t-value shows us.

As for the Cohen's d value, there seems to be little significance or difference between the Name and Numeric/Special Character categories. This restates what the p-value showed as well.

We want to see if there is a significant statistical difference between the mixed and non-mixed categories in relation to the online and offline time cracking. This would allow us to determine if our analysis is valid in the case of different password categories.

```python
# T-tests: Mixed vs Non-Mixed Categories ('password-related' 'simple-alphanumeric' 'nerdy-pop' 'rebellious-rude' vs 'animal' 'sport' 'cool-macho' 'name' 'fluffy' 'food' )

mixed_categories = df[df['category'].str.contains('-')]
non_mixed_categories = df[~df['category'].str.contains('-')]

t_mixed_online, p_mixed_online = stats.ttest_ind(mixed_categories['online_time'], non_mixed_categories['online_time'], equal_var=False)
t_mixed_offline, p_mixed_offline = stats.ttest_ind(mixed_categories['offline_time'], non_mixed_categories['offline_time'], equal_var=False)
```

```python
# Calculate the actual percentage difference

mixed_online_diff = percent_difference(mixed_categories['online_time'], non_mixed_categories['online_time'])
mixed_offline_diff = percent_difference(mixed_categories['offline_time'], non_mixed_categories['offline_time'])

print(f"\nT-Test: Mixed vs Non-Mixed (Online Time) → t = {t_mixed_online:.4f}, p = {p_mixed_online:.4f}, Diff = {mixed_online_diff:.2f}%")
print(f"T-Test: Mixed vs Non-Mixed (Offline Time) → t = {t_mixed_offline:.4f}, p = {p_mixed_offline:.4f}, Diff = {mixed_offline_diff:.2f}%")
```

```python
# Compute Cohen's d effect size
mean_diff = np.mean(mixed_categories['online_time']) - np.mean(non_mixed_categories['online_time'])
pooled_std = np.sqrt((np.std(mixed_categories['online_time'], ddof=1) ** 2 + np.std(non_mixed_categories['online_time'], ddof=1) ** 2) / 2)
cohen_d = mean_diff / pooled_std
print(f'Cohen’s d for Mixed vs Non-Mixed (Online Time): {cohen_d:.3f}')

mean_diff = np.mean(mixed_categories['offline_time']) - np.mean(non_mixed_categories['offline_time'])
pooled_std = np.sqrt((np.std(mixed_categories['offline_time'], ddof=1) ** 2 + np.std(non_mixed_categories['offline_time'], ddof=1) ** 2) / 2)
cohen_d = mean_diff / pooled_std
print(f'Cohen’s d for Mixed vs Non-Mixed (Offline Time): {cohen_d:.3f}')
```

Using the p-value of 0.05 as our threshold, we can see that the p values are much greater than our threshold. From this output, we can see that there is not too much statistical information which supports the notion that there is a statistical difference betwen the categories and online/offline time. As for the t-value, this value would allow us to see the difference between the groups means in relation to variability, similar to standard deviation. In other words, it allows us to see the deviation of the mixed and non-mixed categories in relation to the crack times. Looking at this result, there is minimal variation, around ~0.435 t deviations. Finally, the percent difference value shows us that there isn't too much of a difference between the two categories in terms of the cracking time (around ~11%), restating what the t-value shows us.

As for the Cohen's d value, there seems to be little significance or difference between the Mixed and Non-Mixed categories. This restates what the p-value showed as well.

#### ANOVA tests

This would be valid due to the distributions normal distribution after removal of the outliers, homogeneity of variance, and independence of the observations from each other.

```python
# ANOVA: Does Password Category Affect Cracking Time & Strength?

anova_online = stats.f_oneway(*[df[df['category'] == cat]['online_time'] for cat in df['category'].unique()])
anova_offline = stats.f_oneway(*[df[df['category'] == cat]['offline_time'] for cat in df['category'].unique()])
anova_strength = stats.f_oneway(*[df[df['category'] == cat]['strength'] for cat in df['category'].unique()])

print(f"\nANOVA (Category vs Online Time): F = {anova_online.statistic:.4f}, p = {anova_online.pvalue:.4f}")
print(f"ANOVA (Category vs Offline Time): F = {anova_offline.statistic:.4f}, p = {anova_offline.pvalue:.4f}")
print(f"ANOVA (Category vs Strength): F = {anova_strength.statistic:.4f}, p = {anova_strength.pvalue:.4f}")
```

```python
# Extract groups based on category
groups = [df[df['category'] == cat]['online_time'].values for cat in df['category'].unique()]

# Compute η² (Eta squared) effect size
grand_mean = np.mean(np.concatenate(groups))  # Overall mean
ss_between = sum(len(group) * (np.mean(group) - grand_mean) ** 2 for group in groups)
ss_total = np.var(np.concatenate(groups), ddof=1) * (sum(len(group) for group in groups) - 1)
eta_squared = ss_between / ss_total
print(f'Eta Squared (η²): {eta_squared:.3f}')
```

There seems to be little significance explained by the password category in relation to the strength, and the password online/offline crack time. This is explained by the high p-values in the ANOVA test. In addition, in the Eta squared value, it shows that only 1% of the variance in the password strength and online/offline crack time is explained by the password category.
